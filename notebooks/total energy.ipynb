# Total energy minimization

### import the package

```python
import jrystal
from jrystal.training_utils import create_grids
```

### Load the configuration. You may want to modify yourself.

```python
config = jrystal.config.get_config()
config.grid_sizes = 48
config.epoch = 3000
```

### Perform total energy minimization calculation.

```python
density = jrystal.total_energy.train(config, return_fn="density")
```

### Obtain the unit cell grid.

```python
_, r_vector_grid, _ = create_grids(config)
```

### Evaluate the density at grid points.

```python
density_grid = density(r_vector_grid)
```

### Visualization.

```python
jrystal.visualization.plot_slice(density_grid[0], ncol=24)
```

```python
jrystal.visualization.plot_slice(density_grid[1], ncol=24, same_scale=False)
```
